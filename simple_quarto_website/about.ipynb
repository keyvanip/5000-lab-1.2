```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Sin(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Line Plot')
plt.legend()

plt.grid(True)
plt.show()
```
